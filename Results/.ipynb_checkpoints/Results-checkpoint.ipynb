```python
#Hangover types per group [Above 5 drinks or Below/Equal 5 drinks]
bodies = glob.glob("Body_*.csv")
frame = pd.DataFrame()

above_hangovers=[]
below_hangovers=[]
hangovers=[]

pos_hang=0
for file_ in bodies:
    df_bodies = pd.read_csv(file_, quotechar="'",index_col=None, header=0)
    df_new = df_bodies[df_bodies.Symptom != '-']
    above_hang_sub=[]
    below_hang_sub=[]
    
    for date_ in dates_above[pos_hang]:
        for hang_date in df_new.Date.get_values():
            if (date_ != None and (date_ + timedelta(days=1)).strftime('%Y-%m-%d')==hang_date[0:10]):
                df_hang=df_new[df_new.Date==hang_date]
                above_hang_sub.append(df_hang.Symptom.get_values().tolist())           
    above_hangovers.append(above_hang_sub)
    
    for date1_ in dates_below[pos_hang]:
        for hang_date1 in df_new.Date.get_values():
            if (date1_ != None and (date1_ + timedelta(days=1)).strftime('%Y-%m-%d')==hang_date1[0:10]):
                df_hang1=df_new[df_new.Date==hang_date1]
                below_hang_sub.append(df_hang1.Symptom.get_values().tolist())           
    below_hangovers.append(below_hang_sub)
    pos_hang+=1

print '***************************** Above hangovers ********************************************************************* '
print above_hangovers
print '***************************** Below hangovers ********************************************************************* '
print below_hangovers

    
```

```python
#Frequency of type of hangover per group [Above 5 drinks or Below/Equal 5 drinks]
above_frequency=[0]*10
below_frequency=[0]*10
hang_types=['Perfect - no hangover','Hangover', 'Thirsty','Tired','Headache','Dizziness/faintness','Loss of appetite','Stomach ache','Nausea','Heart racing']
for sub in above_hangovers:
    for ele in sub:
        #print ele
        for t in hang_types: 
            #print t
            if t == ele or t in str(ele):
                above_frequency[hang_types.index(t)]+=1
for sub1 in below_hangovers:
    for ele1 in sub1:
        #print ele
        for t in hang_types: 
            #print t
            if t == ele1 or t in str(ele1):
                below_frequency[hang_types.index(t)]+=1

print 'Above frequency: '
print above_frequency
print 'Number of entries: ', sum(above_frequency)
print 'Below frequency: '
print below_frequency
print 'Number of entries: ',sum(below_frequency)
```

```python
#Percentage of occurrance of type of hangover per group [Above 5 drinks or Below/Equal 5 drinks]
occurrance_above=[]
occurrance_below=[]
for val in above_frequency:
    ocu=float(float(val * 100) / sum(above_frequency))
    occurrance_above.append(round(ocu,2))
print occurrance_above
print sum(occurrance_above)

for val1 in below_frequency:
    ocu1=float(float(val1 * 100) / sum(below_frequency))
    occurrance_below.append(round(ocu1,2))
print occurrance_below
print sum(occurrance_below)
```
